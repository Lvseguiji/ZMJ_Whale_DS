```python
import pandas as pd
import matplotlib.pyplot as plt

women_degree = pd.read_csv('csv/percent-bachelors-degrees-women-usa.csv')
plt.plot(women_degree['Year'], women_degree['Biology'])
plt.show()
```

```python
plt.plot(women_degree['Year'], women_degree['Biology'], c='blue', label='Women')
plt.plot(women_degree['Year'], 100 - women_degree['Biology'], c='red', label='Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded by gender')
plt.show()
```
