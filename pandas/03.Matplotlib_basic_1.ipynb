```python
# 折线图
import pandas as pd
unrate = pd.read_csv('csv/UNRATE.csv')
print(unrate.dtypes)
unrate['DATE'] = pd.to_datetime(unrate['DATE'])  # object->datetime
print(unrate.head(12))
```

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

```python
first_12 = unrate[0:12]  # 前12行数据
plt.plot(first_12.DATE, first_12.VALUE)
plt.show()
```

```python
plt.plot(first_12.DATE, first_12.VALUE)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(first_12.DATE, first_12.VALUE)
plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```
